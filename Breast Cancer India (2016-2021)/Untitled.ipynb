# The data looks just fine so let us jump directly to visualization

```python
#India has 28 states and 8 Union Territories
#lets plot a bar graph for the no of Breast Cancer cases in each state and see which one's the maximum
```

```python
#importing libraries
import matplotlib.pyplot as plt
```

```python
kt = df.copy()
kt.drop(index=36, inplace = True)
kt
```

```python
plt.figure(figsize= (15,10))
plt.bar(kt["States/UT"],kt["2016"])
plt.xlabel("States/UT")
plt.xticks(rotation=90)
plt.ylabel("YEAR 2016")
plt.title("Breast cancer cases in India in 2016")

plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.barh(kt["States/UT"],kt['2016'], color="red")
plt.xticks(rotation=90)
plt.show()
```

```python
#sorting in descending order
df_sorted_desc= kt.sort_values('2016',ascending=False)
```

```python
plt.figure(figsize=(15,10))
kt.sort_values("2016", ascending=True,inplace = True)
plt.barh(kt["States/UT"],kt['2016'], color="red")
plt.xticks(rotation=90)
plt.show()
```

```python
# Now lets print the sorted version of the vertical bar chart 
kt.sort_values('2016', ascending = False, inplace = True)
plt.figure(figsize=(12,7))
plt.bar(kt['States/UT'],kt["2016"])
plt.xlabel("States/UT")
plt.xticks(rotation = 90)
plt.ylabel("2016")
plt.title("Breast cancer cases in India in 2016")
```

```python
for i in kt.columns:
    kt.sort_values(i, ascending = False, inplace = True)
    plt.figure(figsize=(12,7))
    plt.bar(kt['States/UT'],kt[i])
    plt.xlabel("States/UT")
    plt.xticks(rotation = 90)
    plt.ylabel(i)
    plt.title("Breast cancer cases in India in " + i)
```
